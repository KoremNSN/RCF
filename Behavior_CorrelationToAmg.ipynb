```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/media/Data/Lab_Projects/RCF/ReconsolidationOfFea_DATA_2020-06-22_0913.csv')
```

```python
list(df.columns)
```

```python
dfPCL = df[['subject_id','redcap_event_name', 'pcl5_1',
 'pcl5_2',
 'pcl5_3',
 'pcl5_4',
 'pcl5_5',
 'pcl5_6',
 'pcl5_7',
 'pcl5_8',
 'pcl5_9',
 'pcl5_10',
 'pcl5_11',
 'pcl5_12',
 'pcl5_13',
 'pcl5_14',
 'pcl5_15',
 'pcl5_16',
 'pcl5_17',
 'pcl5_18',
 'pcl5_19',
 'pcl5_20']]
```

```python
dfPCL.dropna()
```

```python
dfPCL = dfPCL.rename(columns={'subject_id': 'subject'})
dfPCL = dfPCL.dropna()
dfPCL = dfPCL.astype({"subject": object})
dfPCL
```

```python
sns.distplot(dfPCL['pcl5_1'])
```

```python
# now we need to fix some error in PCL scoring at the VA
# according to Charlie - scores were 0,1,3,4,5 and it should be 0,1,2,3,4
# so we need to fo row by row and in each row column by column of pcl
totalPCL = ['pcl5_1','pcl5_2',  'pcl5_3',  'pcl5_4', 'pcl5_5', 'pcl5_6', 'pcl5_7', 'pcl5_8', 'pcl5_9', 'pcl5_10',
 'pcl5_11', 'pcl5_12', 'pcl5_13', 'pcl5_14', 'pcl5_15', 'pcl5_16', 'pcl5_17', 'pcl5_18', 'pcl5_19', 'pcl5_20']

for i in totalPCL:
    dfPCL[i] = dfPCL[i].mask(dfPCL[i] == 3, 2)
    dfPCL[i] = dfPCL[i].mask(dfPCL[i] == 4, 3)
    dfPCL[i] = dfPCL[i].mask(dfPCL[i] == 5, 4)
```

```python
groupDF = pd.read_excel('/media/Data/Lab_Projects/RCF/PTSD_reconsolidation - to sort/TrioB/Scan_data/newer/rcf_scan_table.xls')
groupDF = groupDF[{'subject_id', 'group'}]
groupDF = groupDF.dropna()
groupDF['subject'] = groupDF['subject_id'].astype(int).astype(str)
groupDF=groupDF.drop(columns = ['subject_id'])
for i in groupDF.iterrows():
    # change two numbers to three
    if len(i[1].subject) < 4:
        groupDF.at[i[0],'subject'] = '0' + i[1].subject
    # add RCF before subject number
    groupDF.at[i[0],'subject'] = 'RCF' + i[1].subject
groupDF.dtypes
```

```python
dfAll = pd.merge(groupDF, dfPCL)
```

```python
dfAll
```

```python
sns.barplot(x = 'group', y= 'pcl5_1', data=dfAll)
```

```python
# lets create a few pcl total scores. 
# total
# anhedonia (PCL12-PCL14)
# avoidance (PCL6 - pcl7)
#dfAll['pcl5_6'] + dfAll['pcl5_7']
totalPCL = ['pcl5_1','pcl5_2',  'pcl5_3',  'pcl5_4', 'pcl5_5', 'pcl5_6', 'pcl5_7', 'pcl5_8', 'pcl5_9', 'pcl5_10',
 'pcl5_11', 'pcl5_12', 'pcl5_13', 'pcl5_14', 'pcl5_15', 'pcl5_16', 'pcl5_17', 'pcl5_18', 'pcl5_19', 'pcl5_20']
avoidance = ['pcl5_6', 'pcl5_7']
anhedonia = ['pcl5_12','pcl5_13', 'pcl5_14']
# sum it up and add to data
dfAll['pcl_Av'] = dfAll[avoidance].sum(axis=1)
dfAll['pcl_An'] = dfAll[anhedonia].sum(axis=1)
dfAll['pcl_total'] = dfAll[totalPCL].sum(axis=1)
dfAll
```

```python
sns.boxplot(x='group',y='pcl_total', data=dfAll)
```

```python
sns.boxplot(x='group',y='pcl_An', data=dfAll)
```
